```python
import numpy as np
import matplotlib.pyplot as plt

# Given points
x_values = np.array([0, 50])
y_values = np.array([1, 0.4])

# Fit a polynomial of degree n to the data points
n = 3
coefficients = np.polyfit(x_values, y_values, n)

# Define a function representing the polynomial equation
polynomial_function = np.poly1d(coefficients)

# Define a function to find the root (intersection with x-axis) using Newton's method
def find_root_newton(f, f_prime, x_initial, tol=1e-6, max_iter=1000):
    """Find root of function f using Newton's method."""
    x = x_initial
    iterations = 0
    while iterations < max_iter:
        x_next = x - f(x) / f_prime(x)
        if abs(x_next - x) < tol:
            return x_next
        x = x_next
        iterations += 1
    print("Maximum iterations reached. Root not found.")
    return None

# Define the derivative of the polynomial function
polynomial_derivative = polynomial_function.deriv()

# Find the root (intersection with x-axis) using Newton's method
root = find_root_newton(polynomial_function, polynomial_derivative, x_initial=x_values[0])

if root:
    print("The curve intersects the x-axis at x =", root)
else:
    print("No root found within the specified interval.")
print("The curve intersects the x-axis at x =", root)

# Generate x values for plotting
x_plot = np.linspace(min(x_values), max(x_values), 100)
# Compute corresponding y values using the polynomial function
y_plot = polynomial_function(x_plot)

x_plot_extended = np.linspace(min(x_values), root, 100)
# Compute corresponding y values using the polynomial function for the extended range
y_plot_extended = polynomial_function(x_plot_extended)

# Plot the given points, the interpolated curve, and the intersection point with x-axis
plt.scatter(x_values, y_values, color='red', label='Given points')
plt.plot(x_plot_extended, y_plot_extended, color='blue', label='Interpolated curve')
plt.scatter(root, 0, color='green', label=f'Intersection with x-axis at x={root:.2f}')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Interpolation with Extended Curve')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate the Remaining Useful Life (RUL)
RUL = root - x_values[-1]  # Subtract the initial x value from the root value

# Print the RUL
print("Remaining Useful Life (RUL):", RUL)
```
