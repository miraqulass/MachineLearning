```python
for col in num_cols:
  print(col)
  print('Skew:', round(data[col].skew(), 2))
  plt.figure(figsize = (15, 4))
  plt.subplot(1, 2, 1)
  data[col].hist(grid=False)
  plt.ylabel('count')
  plt.subplot(1, 2, 2)
  sns.boxplot(x=data[col])
  plt.show()
```

```python
fig, axes = plt.subplots(3, 2, figsize=(15, 10))
fig.suptitle('Bar plots for all the Categorical Variables')
sns.countplot(ax = axes[0,0], x='GPS', data=data, color='blue',
              order=data['GPS'].head(20).value_counts().index);

sns.countplot(ax = axes[0,1], x='MaintenanceType', data=data, color='blue',
              order=data['MaintenanceType'].head(20).value_counts().index);

sns.countplot(ax = axes[1,0], x='MaintenanceCategory', data=data, color='blue',
              order=data['MaintenanceCategory'].head(20).value_counts().index);

sns.countplot(ax = axes[1,1], x='MaintenanceFrequency', data=data, color='blue',
              order=data['MaintenanceFrequency'].head(20).value_counts().index);

sns.countplot(ax = axes[2,0], x='MaintenanceSeverity', data=data, color='blue',
              order=data['MaintenanceSeverity'].head(20).value_counts().index);

sns.countplot(ax = axes[2,1], x='MaintenanceDate', data=data, color='blue',
              order=data['MaintenanceDate'].head(20).value_counts().index);

sns.countplot(ax=axes[1, 1], x='MaintenanceInterval', data=data, color='blue',
              order=data['MaintenanceInterval'].head(20).value_counts().index);

axes[0,0].tick_params(labelrotation=90)
axes[0,1].tick_params(labelrotation=90)
# axes[1,0].tick_params(labelrotation=90)
axes[1,1].tick_params(labelrotation=90)
# axes[2,0].tick_params(labelrotation=90)
axes[2,1].tick_params(labelrotation=90)
# axes[1,1].tick_params(labelrotation=90)
```
